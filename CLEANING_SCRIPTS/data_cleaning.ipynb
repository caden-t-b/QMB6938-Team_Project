```python
import pandas as pd
from IPython.display import display
```

```python
base_df = pd.read_csv("../RAW_DATA/flinfo2022.csv")
general_stats_df = pd.read_csv("../RAW_DATA/flhealthcharts_summary_2021_wide_snake.csv")
main_pct_df = pd.read_csv("../RAW_DATA/flhealthcharts_model_ready_full.csv")

display(base_df)
display(general_stats_df)
display(main_pct_df)
```

```python
base_df = base_df.rename(columns={'Admin2': 'county'})
base_df = base_df[base_df['county'] != 'Unassigned']
base_df.head()
```

```python
# Get unique counties from each dataframe
counties_master = set(base_df['county'].unique())
counties_general = set(general_stats_df['county'].unique())
counties_main = set(main_pct_df['county'].unique())

# Counties in all three
common_to_all = counties_master & counties_general & counties_main
print("Counties in all three dataframes:")
print(common_to_all)

# Counties only in base_df
only_master = counties_master - counties_general - counties_main
print("\nCounties only in base_df:")
print(only_master)

# Counties only in general_stats_df
only_general = counties_general - counties_master - counties_main
print("\nCounties only in general_stats_df:")
print(only_general)

# Counties only in main_pct_df
only_main = counties_main - counties_master - counties_general
print("\nCounties only in main_pct_df:")
print(only_main)
```

```python
"""
master_df = base_df[["county", "Case_Fatality_Ratio"]].merge(
    general_stats_df, on="county", how="inner"
).merge(
    main_pct_df, on="county", how="inner"
)
"""
#master_df = base_df[["county", "Case_Fatality_Ratio"]].merge(main_pct_df, on="county", how="inner")
master_df = base_df[["county", "Case_Fatality_Ratio"]].merge(general_stats_df, on="county", how="inner")

master_df = master_df.dropna(axis=1, how="any")

master_df.head()
```

```python
import numpy as np
import pandas as pd

def drop_collinear_features(df, threshold=0.9):
    corr_matrix = df.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    return df.drop(columns=to_drop), to_drop
```

```python
# Separate features & target
X = master_df.drop(columns=["county", "Case_Fatality_Ratio"])
y = master_df["Case_Fatality_Ratio"]

# Drop collinear features
X_reduced, dropped_features = drop_collinear_features(X, threshold=0.9)

print("Dropped features due to collinearity:", dropped_features)
print("Remaining features:", X_reduced.shape[1])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

X = master_df.drop(columns=["Case_Fatality_Ratio", "county"])
y = master_df["Case_Fatality_Ratio"]

# Ridge regression
ridge = Pipeline([
    ("scaler", StandardScaler()),
    ("model", Ridge(alpha=1.0))
])
ridge_scores = cross_val_score(ridge, X, y, cv=5, scoring="r2")
print("Ridge RÂ²:", np.mean(ridge_scores))

# Lasso regression
lasso = Pipeline([
    ("scaler", StandardScaler()),
    ("model", Lasso(alpha=0.01))
])
lasso_scores = cross_val_score(lasso, X, y, cv=5, scoring="r2")
print("Lasso RÂ²:", np.mean(lasso_scores))

# Random Forest
rf = RandomForestRegressor(n_estimators=500, random_state=1)
rf_scores = cross_val_score(rf, X, y, cv=5, scoring="r2")
print("Random Forest RÂ²:", np.mean(rf_scores))
```
